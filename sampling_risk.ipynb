# Calculate the credible interval of sampling items (instead of testing all items)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels as sm
```

```python
from empiricaldist import Pmf
```

```python
from utils import decorate
```

Imagine 11 programmers all identical to me except that the errors they introduce in implementing features ranges from every time to never. Alternatively, the percentage of error free code they write ranges from 0% to 100%. These programmers are our hypotheses and their priors.

```python
hypotheses = np.arange(0, 100 + 1, 10)  # ranges from zero to 100 by tens
prior = Pmf(1, hypotheses)
prior.normalize()
prior
```

We use the likelihood from the "cookie problem" in "Think Bayes", 2nd edition.,

```python
likelihood_pass = hypotheses / 100
```

```python
passing_1 = prior * likelihood_pass
passing_1.normalize()
```

We "steal" the `decorate_bowls` function from "Think Bayes", 2nd edition.

```python
def decorate_bowls(title):
    decorate(xlabel='Bowl #',
             ylabel='PMF',
             title=title)
```

```python
prior.plot(label='prior', color='C5')
passing_1.plot(label='posterior', color='C4')
decorate_bowls('Posterior after one passing sample')
```

```python
passing_1.credible_interval(0.95)
```

And now we just "turn the crank"

```python
passing_2 = passing_1 * likelihood_pass
```

```python
prior.plot(label='prior', color='C5')
passing_1.plot(label='Passing 1 sample', color='C4')
passing_2.plot(label='Passing 2 Samples', color='C3')
decorate_bowls('Posterior after two passing samples')
```
